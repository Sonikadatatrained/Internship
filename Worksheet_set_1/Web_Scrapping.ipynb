```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm ')
html_content = page.content
soup = BeautifulSoup(html_content, "html.parser")
table = soup.find("table")

if table:
    rows = table.find_all("tr")[1:]
    president_names = []
    president_terms = []
    for row in rows:
        columns = row.find_all("td")
        name = columns[0].get_text(strip=True)
        term = columns[1].get_text(strip=True)
        president_names.append(name)
        president_terms.append(term)
        data = {
        "Name": president_names,"Term of Office": president_terms
                }
        df = pd.DataFrame(data)
        print(df)
    else:
        print("Table not found")
```

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
articles = soup.find_all("div", class_="Card-titleContainer")
headlines = []
times = []
news_links = []
for article in articles:
    headline = article.find("a", class_="Card-title").get_text(strip=True)
    time_element = article.find("time")
    if time_element:
        time = time_element.get_text(strip=True)
    else:
        time = "N/A"
    

    time = article.find("time").get_text(strip=True) if time_element else "N/A"
    news_link = article.find("a", class_="Card-title")["href"]
    
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)
data = {
    "Headline": headlines,
    "Time": times,
    "News Link": news_links
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
articles = soup.find_all("div", class_="pod-listing-header")
paper_titles = []
authors_list = []
published_dates = []
paper_urls = []

for article in articles:
    title_element = article.find("a", class_="pod-listing-header__link")
    paper_title = title_element.get_text(strip=True)
    paper_url = "https://www.journals.elsevier.com" + title_element["href"]
    
    authors_element = article.find("div", class_="pod-listing-header__authors")
    authors = authors_element.get_text(strip=True)
    
    date_element = article.find("div", class_="pod-listing-header__date")
    published_date = date_element.get_text(strip=True)
    
    paper_titles.append(paper_title)
    authors_list.append(authors)
    published_dates.append(published_date)
    paper_urls.append(paper_url)
    
    data = {
    "Paper Title": paper_titles,
    "Authors": authors_list,
    "Published Date": published_dates,
    "Paper URL": paper_urls
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
url = "https://www.dineout.co.in/delhi-restaurants"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content, "html.parser")
restaurants = soup.find_all("div", class_="restnt-card")
restaurant_names = []
cuisines = []
locations = []
ratings = []
image_urls = []
for restaurant in restaurants:
    name_element = restaurant.find("div", class_="restnt-card__descriptive--name")
    restaurant_name = name_element.get_text(strip=True) if name_element else "N/A"
    
    cuisine_element = restaurant.find("div", class_="restnt-card__descriptive--cuisine")
    cuisine = cuisine_element.get_text(strip=True) if name_element else "N/A"
    
    location_element = restaurant.find("div", class_="restnt-card__descriptive--location")
    location = location_element.get_text(strip=True) if name_element else "N/A"
    
    rating_element = restaurant.find("span", class_="restnt-card__rating--val")
    rating = rating_element.get_text(strip=True) if rating_element else "N/A"
    
    image_element = restaurant.find("img", class_="restnt-card__img--img")
    image_url = image_element["data-src"] if image_element else "N/A"
    
    restaurant_names.append(restaurant_name)
    cuisines.append(cuisine)
    locations.append(location)
    ratings.append(rating)
    image_urls.append(image_url)
    data = {
    "Restaurant Name": restaurant_names,
    "Cuisine": cuisines,
    "Location": locations,
    "Ratings": ratings,
    "Image URL": image_urls
}
df = pd.DataFrame(data)
df
```
